# Intro
Say you have generated new autoTST reaction objects and you want to add them to the database of known reactions as well as update the species dictionary for any unknown species in these new reactions:

```python
"""
Getting some test reactions to use as examples.
"""

from autotst.reaction import AutoTST_Reaction
test_labels = ['C+[CH2]COC(=O)C(C)C_CCOC(=O)C(C)C+[CH3]',
 'O+[CH2]C(C)CC(C)C(C)=O_CC(=O)C(C)CC(C)C+[OH]',
 'O+[CH2]C(C)OC(CC)CC_CCC(CC)OC(C)C+[OH]',
 'O+[CH2]COC(=O)C(C)C_CCOC(=O)C(C)C+[OH]',
 'O+[CH]=CC(C)=CCCC(=C)C_C=CC(C)=CCCC(=C)C+[OH]',
 'OO+[CH2]C(=C)C(=O)CC_C=C(C)C(=O)CC+[O]O',
 'OO+[CH2]C(=C)CC(C)=O_C=C(C)CC(C)=O+[O]O',
 'OO+[CH2]C(=C)CCC=C(C)C=C_C=CC(C)=CCCC(=C)C+[O]O',
 'OO+[CH2]C(=O)C(=C)C_C=C(C)C(C)=O+[O]O',
 'OO+[CH2]C(=O)C(C)(C)C_CC(=O)C(C)(C)C+[O]O']

test_reactions = [AutoTST_Reaction(label, reaction_family='H_Abstraction') for label in test_labels]
```

```python
# Getting the new stuff
import autotst.updater_methods as UPmethods
```

```python
# some kwargs:
method = 'm062x/6-311+G(2df,2p)'
shortDesc = 'M06-2X/6-311+G(2df,2p) calculation via group additive TS generator.'
# if no family given, will attempt as H abstraction
family = 'H_Abstraction'
UPmethods.update_databases(test_reactions, method=method, shortDesc=shortDesc, reaction_family=family)
#updater_methods.update_databases?
```

Now the test reactions (and all previous in reactions.py) have been added to 'updated_reactions.py' and all new species have been combined with exisitng species in 'updated_dictionary.txt'.

If you want to change any of the paths involved with loading previous databases or where it is saving these updates, they can all be found in updater_method.update_databases().

Right now the file names are set up as to not overwrite the existing files out of caution.

# Now to update the TS_groups.py file using our new reactions.py and dictionary.txt

Keep in mind this process will use 'reactions.py' and 'dictionary.txt' so without overwriting them in the previous step or specifically indication the path to the updated files this will not actually update anything 

```python
# Provide list of families you wish to update
families = ['H_Abstraction']

# Call this method to recieve a dictionary of Updater Instances (which themselves carry the updated databases and info)
family_instances = UPmethods.TS_Database_Update(families)
#updater_methods.TS_Database_Update?
```

Updated databases are not yet saved. But can be accessed through their respective updater instances

```python
h_ab_updater_instance = family_instances['H_Abstraction']

training_set = h_ab_updater_instance.training_set
database = h_ab_updater_instance.database

database.groups.entries
```
